```python
import math

def parse_currency(value):
    if isinstance(value, str):
        return float(value.replace('$', '').replace(',', '').strip())
    return float(value)

def insertion_sort(arr, low, high):
    for i in range(low + 1, high + 1):
        key_item = arr[i]
        j = i - 1
        while j >= low and arr[j][1] < key_item[1]:  # Descending sort
            arr[j + 1] = arr[j]
            j -= 1
        arr[j + 1] = key_item

def heapify(arr, n, i):
    largest = i
    left = 2 * i + 1
    right = 2 * i + 2

    if left < n and arr[left][1] > arr[largest][1]:
        largest = left
    if right < n and arr[right][1] > arr[largest][1]:
        largest = right

    if largest != i:
        arr[i], arr[largest] = arr[largest], arr[i]
        heapify(arr, n, largest)

def heap_sort(arr):
    n = len(arr)
    for i in range(n // 2 - 1, -1, -1):
        heapify(arr, n, i)
    for i in range(n - 1, 0, -1):
        arr[0], arr[i] = arr[i], arr[0]
        heapify(arr, i, 0)

def partition(arr, low, high):
    pivot = arr[high][1]
    i = low - 1
    for j in range(low, high):
        if arr[j][1] > pivot: 
            i += 1
            arr[i], arr[j] = arr[j], arr[i]
    arr[i + 1], arr[high] = arr[high], arr[i + 1]
    return i + 1

def intro_sort_util(arr, low, high, maxdepth):
    if high - low <= 16:
        insertion_sort(arr, low, high)
    elif maxdepth == 0:
        temp = arr[low:high + 1]
        heap_sort(temp)
        arr[low:high + 1] = temp[::-1]  
    else:
        pivot = partition(arr, low, high)
        intro_sort_util(arr, low, pivot - 1, maxdepth - 1)
        intro_sort_util(arr, pivot + 1, high, maxdepth - 1)

def intro_sort(arr):
    maxdepth = int(2 * math.log2(len(arr))) if len(arr) > 0 else 0
    intro_sort_util(arr, 0, len(arr) - 1, maxdepth)
    return arr

def top_profitable_products(data):
    product_profit = {}
    for row in data:
        product = row['Product Name']
        profit = parse_currency(row['Profit'])
        if product in product_profit:
            product_profit[product] += profit
        else:
            product_profit[product] = profit

    product_profit_list = list(product_profit.items())

    sorted_products = intro_sort(product_profit_list)

    return sorted_products[:10]

top_products = top_profitable_products(data)
print("Top 10 Profitable Products:")
for product, profit in top_products:
    print(f"{product}: ${profit:.2f}")
```

# DYNAMIC ANALYSIS

```python
import random
import time
import sys

sys.setrecursionlimit(10000)

def selection_sort(arr):
    n = len(arr)
    for i in range(n):
        min_idx = i
        for j in range(i + 1, n):
            if arr[j][1] < arr[min_idx][1]:
                min_idx = j
        arr[i], arr[min_idx] = arr[min_idx], arr[i]
    return arr

def merge_sort(arr):
    if len(arr) <= 1:
        return arr
    mid = len(arr) // 2
    left = merge_sort(arr[:mid])
    right = merge_sort(arr[mid:])
    return merge(left, right)

def merge(left, right):
    result = []
    i = j = 0
    while i < len(left) and j < len(right):
        if left[i][1] < right[j][1]:
            result.append(left[i])
            i += 1
        else:
            result.append(right[j])
            j += 1
    result.extend(left[i:])
    result.extend(right[j:])
    return result

def introsort(arr):
    max_depth = (len(arr).bit_length() - 1) * 2
    _introsort_helper(arr, 0, len(arr) - 1, max_depth)
    return arr

def _introsort_helper(arr, start, end, max_depth):
    if end - start <= 1:
        return
    elif max_depth == 0:
        heapsort(arr, start, end)
    else:
        pivot = partition(arr, start, end)
        _introsort_helper(arr, start, pivot - 1, max_depth - 1)
        _introsort_helper(arr, pivot + 1, end, max_depth - 1)

def partition(arr, low, high):
    pivot = arr[high][1]
    i = low
    for j in range(low, high):
        if arr[j][1] < pivot:
            arr[i], arr[j] = arr[j], arr[i]
            i += 1
    arr[i], arr[high] = arr[high], arr[i]
    return i

def heapsort(arr, start, end):
    def heapify(arr, n, i):
        largest = i
        l = 2 * i + 1 - start
        r = 2 * i + 2 - start
        if l <= end - start and arr[start + l][1] > arr[start + largest][1]:
            largest = l
        if r <= end - start and arr[start + r][1] > arr[start + largest][1]:
            largest = r
        if largest != i:
            arr[start + i], arr[start + largest] = arr[start + largest], arr[start + i]
            heapify(arr, n, largest)

    n = end - start + 1
    for i in range(n // 2 - 1, -1, -1):
        heapify(arr, n, i)
    for i in range(n - 1, 0, -1):
        arr[start], arr[start + i] = arr[start + i], arr[start]
        heapify(arr, i, 0)

def generate_data(size):
    return [(f"Product{i}", random.uniform(0, 10000)) for i in range(size)]

def time_algorithm(algorithm, data):
    copied_data = data.copy()
    start = time.time()
    algorithm(copied_data)
    end = time.time()
    return round((end - start) * 1000, 2)  

sizes = [500, 1000, 2000, 3000, 4000, 5000]
results = {}

for size in sizes:
    test_data = generate_data(size)
    results[size] = {
        "Selection Sort": time_algorithm(selection_sort, test_data),
        "Merge Sort": time_algorithm(merge_sort, test_data),
        "IntroSort": time_algorithm(introsort, test_data),
    }

for size, timings in results.items():
    print(f"\nSize: {size}")
    for algo, time_taken in timings.items():
        print(f"{algo}: {time_taken} ms")
```

```python
def selection_sort(arr):                # O(n)
    n = len(arr)
    for i in range(n):                  # O(n)
        min_idx = i
        for j in range(i + 1, n):       # O(n) 
            if arr[j][1] < arr[min_idx][1]:  # O(1)
                min_idx = j
        arr[i], arr[min_idx] = arr[min_idx], arr[i]  # O(1)
    return arr
```

```python
def merge_sort(arr):                   # T(n)
    if len(arr) <= 1:                  # O(1)
        return arr
    mid = len(arr) // 2
    left = merge_sort(arr[:mid])      # T(n/2)
    right = merge_sort(arr[mid:])     # T(n/2)
    return merge(left, right)         # O(n)

def merge(left, right):
    result = []
    i = j = 0
    while i < len(left) and j < len(right):  # O(n)
        if left[i][1] < right[j][1]:
            result.append(left[i])
            i += 1
        else:
            result.append(right[j])
            j += 1
    result.extend(left[i:])           # O(n)
    result.extend(right[j:])          # O(n)
    return result
```

```python
def introsort(arr):
    max_depth = (len(arr).bit_length() - 1) * 2  # O(1)
    _introsort_helper(arr, 0, len(arr) - 1, max_depth)
    return arr                                   # O(n log n)

def _introsort_helper(arr, start, end, max_depth):
    if end - start <= 1:                # O(1)
        return
    elif max_depth == 0:
        heapsort(arr, start, end)       #O(n log n)
    else:
        pivot = partition(arr, start, end)       # O(n)
        _introsort_helper(arr, start, pivot - 1, max_depth - 1)  # O(log n) 
        _introsort_helper(arr, pivot + 1, end, max_depth - 1)

def partition(arr, low, high):
    pivot = arr[high][1]               # O(1)
    i = low
    for j in range(low, high):         # O(n)
        if arr[j][1] < pivot:          # O(1)
            arr[i], arr[j] = arr[j], arr[i]  # O(1)
            i += 1
    arr[i], arr[high] = arr[high], arr[i]    # O(1)
    return i

def heapsort(arr, start, end):
    def heapify(arr, n, i):
        largest = i
        l = 2 * i + 1 - start          # O(1)
        r = 2 * i + 2 - start
        if l <= end - start and arr[start + l][1] > arr[start + largest][1]:
            largest = l
        if r <= end - start and arr[start + r][1] > arr[start + largest][1]:
            largest = r
        if largest != i:
            arr[start + i], arr[start + largest] = arr[start + largest], arr[start + i]  # O(1)
            heapify(arr, n, largest)  # O(log n)

    n = end - start + 1
    for i in range(n // 2 - 1, -1, -1):  # O(n)
        heapify(arr, n, i)
    for i in range(n - 1, 0, -1):        # O(n log n)
        arr[start], arr[start + i] = arr[start + i], arr[start]
        heapify(arr, i, 0)
```

```python
def selection_sort(items):
    assert isinstance(items, list), "Input must be a list"
    
    assert all(isinstance(i, tuple) for i in items), "All elements must be tuples"
    
    assert all(isinstance(i[1], (int, float)) for i in items), "Second item in tuple must be numeric"

    n = len(items)
    for i in range(n):
        max_index = i
        for j in range(i + 1, n):
            if items[j][1] > items[max_index][1]: 
                max_index = j
        items[i], items[max_index] = items[max_index], items[i]
    

    for k in range(len(items) - 1):
        assert items[k][1] >= items[k + 1][1], "List is not sorted in descending order"
    
    return items
```

```python
def merge_sort(items):
    assert isinstance(items, list), "Input must be a list"
    assert all(isinstance(i, tuple) for i in items), "Each item must be a tuple"

    if len(items) <= 1:
        return items
    mid = len(items) // 2
    left = merge_sort(items[:mid])
    right = merge_sort(items[mid:])
    return merge(left, right)

def merge(left, right):
    result = []
    i = j = 0
    while i < len(left) and j < len(right):
        assert isinstance(left[i], tuple) and isinstance(right[j], tuple), "Elements must be tuples"
        if left[i][1] > right[j][1]:  # Descending sort
            result.append(left[i])
            i += 1
        else:
            result.append(right[j])
            j += 1
    result.extend(left[i:])
    result.extend(right[j:])
    return result
```

```python
def _introsort_helper(arr, start, end, max_depth):

    assert isinstance(arr, list), "arr must be a list"
    assert isinstance(start, int) and isinstance(end, int), "start and end must be integers"
    assert start <= end, "start index must be <= end index"

    if end - start <= 1:
        return

    elif max_depth == 0:
        heapsort(arr, start, end)

    else:
        pivot = partition(arr, start, end)
        _introsort_helper(arr, start, pivot - 1, max_depth - 1)
        _introsort_helper(arr, pivot + 1, end, max_depth - 1)
```

```python
def partition(arr, low, high):
    assert isinstance(arr, list)
    assert isinstance(low, int) and isinstance(high, int), "low and high should be integers"
    assert low <= high, "low should be less than or equal to high"
    
    pivot = arr[high][1] 
    i = low
    
    for j in range(low, high):
        assert isinstance(arr[j], tuple), "arr elements should be tuples"
        assert len(arr[j]) > 1, "tuples should have at least 2 elements"
        if arr[j][1] < pivot:
            arr[i], arr[j] = arr[j], arr[i] 
            i += 1
    
    arr[i], arr[high] = arr[high], arr[i] 
    
    assert low <= i <= high, "Returned index i out of range"
    return i
```
